```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Počet trojuholníkov

```python
x = []
y = []
for i in range(0,221):
    vertices = []
    faces = []
    text = '000'+str(i)
    with open('frame'+text[-4:]+'.obj', 'r') as file:
        file_contents = file.read()
        count = file_contents.count('f')
    x.append(i)
    y.append(count)
    
for i in range(1,len(x)):
    if (y[i] - y[i-1])!=0:
        zlom = i
        break
print(zlom)
plt.plot(x,y)
plt.title('Počet trojuholníkov pre každú iteráciu')
plt.show()
```

Grafy vývoja priemerov mier a osbahov

```python
df1 = pd.read_csv('csv/data1.csv')
df2 = pd.read_csv('csv/data2.csv')
```

```python
plt.plot(df1['Index'],df1['pomer_polomerov'])
plt.plot(df1['Index'],df1['pomer_extrem_stran'])
plt.plot(df1['Index'],df1['pomer_polomer_polobvod'])
plt.plot(df1['Index'],df1['pomer_stran'])
plt.plot(df1['Index'],df1['pomer_opisana_strana'])
plt.title("Vývoj priemerov mier")
plt.legend(['pomer_polomerov','pomer_extrem_stran','pomer_polomer_polobvod','pomer_stran','pomer_opisana_strana',])

plt.show()

plt.plot(df1['Index'], df1['obsahy'])
plt.title('Vývoj priemerného obsahu')
plt.show()
```

Grafy vývoja mediánov mier a obsahov

```python
plt.plot(df2['Index'],df2['median pomer_polomerov'])
plt.plot(df2['Index'],df2['median pomer_extrem_stran'])
plt.plot(df2['Index'],df2['median pomer_polomer_polobvod'])
plt.plot(df2['Index'],df2['median pomer_stran'])
plt.plot(df2['Index'],df2['median pomer_opisana_strana'])
plt.title("Vývoj miedánov mier")
plt.legend(['pomer_polomerov','pomer_extrem_stran','pomer_polomer_polobvod','pomer_stran','pomer_opisana_strana',])

plt.show()

plt.plot(df2['Index'], df2['median obsahy'])
plt.title('Vývoj medianového obsahu')
plt.show()
```

Grafy ktorých porovnávame ako sa líši priemer od mediánu pre miery a obsahy

```python
plt.title("Vývoj rozdielu priemeru a mediánu mier")
plt.plot(df1['Index'],np.abs(df1['pomer_polomerov'] - df2['median pomer_polomerov']))
plt.plot(df1['Index'],np.abs(df1['pomer_extrem_stran'] - df2['median pomer_extrem_stran']))
plt.plot(df1['Index'],np.abs(df1['pomer_polomer_polobvod'] - df2['median pomer_polomer_polobvod']))
plt.plot(df1['Index'],np.abs(df1['pomer_stran'] - df2['median pomer_stran']))
plt.plot(df1['Index'],np.abs(df1['pomer_opisana_strana'] - df2['median pomer_opisana_strana']))
plt.legend(['pomer_polomerov','pomer_extrem_stran','pomer_polomer_polobvod','pomer_stran','pomer_opisana_strana',])

plt.show()
plt.title("Vývoj rozdielu priemeru a mediánu obsahu")
plt.plot(df1['Index'], np.abs(df1['obsahy'] - df2['median obsahy']))
plt.show()
```
